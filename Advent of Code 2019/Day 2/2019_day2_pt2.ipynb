```python
def run_program(noun, verb, data):
    data[1] = noun
    data[2] = verb
    pos = 0
    while data[pos] != 99:
        if data[pos] == 1:
            p1 = data[data[pos + 1]]
            p2 = data[data[pos + 2]]
            data[data[pos + 3]] = p1 + p2
        if data[pos] == 2:
            p1 = data[data[pos + 1]]
            p2 = data[data[pos + 2]]
            data[data[pos + 3]] = p1 * p2
        pos += 4
    return data[0]
```

```python
for noun in range(100):
    for verb in range(100):
        total = run_program(noun, verb, d.copy())
        print(total)
        if total == 19690720:
            break
    if total == 19690720:
        break
```
