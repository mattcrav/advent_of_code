```python
import pandas as pd
import numpy as np
```

```python
# df = pd.read_csv('2022_day4_sample.csv', header=None, names=['left', 'right'], delimiter=',')
df = pd.read_csv('2022_day4.csv', header=None, names=['left', 'right'], delimiter=',')
df.head()
```

# Set Method

Did not work at first because df['both'] was not sorted

```python
df['leftr'] = df['left'].apply(lambda x: list(range(int(x.split('-')[0]), int(x.split('-')[1]) + 1)))
df['rightr'] = df['right'].apply(lambda x: list(range(int(x.split('-')[0]), int(x.split('-')[1]) + 1)))
df.head(10)
```

```python
df['both'] = df.apply(lambda x: list((set(x['leftr'])) & (set(x['rightr']))), 1)
df['both'].apply(lambda x: x.sort())
df.head(10)
```

```python
df['fully'] = np.where((df['both'] == df['leftr']) | (df['both'] == df['rightr']), 1, 0)
df.head(10)
```

# Comparitive Method

```python
df['left_1'] = df['left'].str.split('-').apply(lambda x: int(x[0]))
df['left_2'] = df['left'].str.split('-').apply(lambda x: int(x[1]))
df['right_1'] = df['right'].str.split('-').apply(lambda x: int(x[0]))
df['right_2'] = df['right'].str.split('-').apply(lambda x: int(x[1]))
df.head(10)
```

```python
df['left_fully'] = np.where((df['left_1'] >= df['right_1']) & (df['left_2'] <= df['right_2']), 1, 0)
df['right_fully'] = np.where((df['right_1'] >= df['left_1']) & (df['right_2'] <= df['left_2']), 1, 0)
df['fully_2'] = np.maximum(df['left_fully'], df['right_fully'])
df.head(10)
```
