# Directory names are not unique! Need the whole path

```python
import pandas as pd
import numpy as np
```

```python
b = pd.read_csv('2022_day7.csv', header=None)
# b = pd.read_csv('2022_day7_sample.csv', header=None)
b.head(25)
```

```python
f = b[(b[0] != '$ ls') & (~b[0].str.contains('dir '))].copy()
fs = f[0].to_list()
f.head(25)
```

```python
d = ''
sz = {}
up = []
for c in fs:
    if '$' in c:
        v = c.replace('$ cd ', '')
        if v == '..':
            up.pop()
            d = up[-1]
        else:
            d = d + v + '/'
            sz[d] = 0
            up.append(d)
        # print(c, up)
    else:
        size = int(c.split(' ')[0])
        for u in up:
            sz[u] += size
print(sz)
```

```python
ts = [x for x in sz.values() if x <= 100000]
print(ts)
print(sum(ts))
```
