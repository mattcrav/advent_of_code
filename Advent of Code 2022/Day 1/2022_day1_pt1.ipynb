```python
# with open('2022_day1_sample.csv', 'r') as f:
#     l = f.readlines()
with open('2022_day1.csv', 'r') as f:
    l = f.readlines()
print(l)
```

```python
g = ''.join(l).split('\n\n')
print(g)
```

```python
gl = [[int(y) for y in x.split('\n')] for x in g]
print(gl)
```

```python
s = [sum(x) for x in gl]
print(s)
print(max(s))
```
