```python
with open('2015_day3.csv', 'r') as f:
    txt = f.read()
```

```python
current_pos = [0, 0]
posid = str(current_pos[0]) + ',' + str(current_pos[1])
visited = {posid: 1}
```

```python
for d in txt:
    if d == '^':
        current_pos[0] += 1
    if d == '>':
        current_pos[1] += 1
    if d == '<':
        current_pos[1] -= 1
    if d == 'v':
        current_pos[0] -= 1
    # print(d, current_pos)
    posid = str(current_pos[0]) + ',' + str(current_pos[1])
    if posid not in visited:
        visited[posid] = 1
    else:
        visited[posid] += 1
```

```python
print(len(visited))
```
