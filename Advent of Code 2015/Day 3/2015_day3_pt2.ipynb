```python
with open('2015_day3.csv', 'r') as f:
    txt = f.read()
```

```python
santa_pos = [0, 0]
robot_pos = [0, 0]
posid = str(santa_pos[0]) + ',' + str(santa_pos[1])
visited = {posid: 2}
```

```python
counter = 0
for d in txt:
    if counter % 2 == 0:
        pos = santa_pos
    else:
        pos = robot_pos
    if d == '^':
        pos[0] += 1
    if d == '>':
        pos[1] += 1
    if d == '<':
        pos[1] -= 1
    if d == 'v':
        pos[0] -= 1
    # print(d, current_pos)
    posid = str(pos[0]) + ',' + str(pos[1])
    if posid not in visited:
        visited[posid] = 1
    else:
        visited[posid] += 1
    counter += 1
```
