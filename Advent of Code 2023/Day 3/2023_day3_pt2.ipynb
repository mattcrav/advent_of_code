```python
import numpy as np
from collections import defaultdict
```

```python
with open('2023_day3.csv', 'r') as f:
    s = f.readlines()
s = [list(x.replace('\n','')) for x in s]
a = np.array(s)
print(a)
```

```python
# symbol locations
f = np.where(a == '*', 1, 0)
f = f.astype(int)
i = 1
# number the gear locations to we tell which numbers share the same gear
for r in range(len(f)):
    for c in range(len(f[r])):
        if f[r, c] == 1:
            f[r, c] = i
            i += 1
print(f)
```

```python
# valid part num locations
h = f + np.roll(f, 1, axis=0) # down
h = h + np.roll(f, 1, axis=1) # right
h = h + np.roll(f, -1, axis=0) # up
h = h + np.roll(f, -1, axis=1) # left
h = h + np.roll(np.roll(f, 1, axis=0), -1, axis=1) # lower left
h = h + np.roll(np.roll(f, 1, axis=0), 1, axis=1) # lower right
h = h + np.roll(np.roll(f, -1, axis=0), -1, axis=1) # upper left
h = h + np.roll(np.roll(f, -1, axis=0), 1, axis=1) # upper right
h = np.where(h == 0, 0, h)
print(h)
```

```python
# number locations
# don't want to confuse actual zeroes with . or symbols so set to 10
n = np.where(a == '0', 10, a)
n = np.where(np.isin(a, [str(x) for x in range(10)]), n, 0)
n = n.astype(int)
print(n)
```

```python
# locations where a digit is valid part number because adjacent to symbol
v = h * n
print(v)
```

```python
# return actual zeroes from 10 to 0
nt = np.where(n == 0, -1, n)
nt = np.where(n == 10, 0, nt)
print(nt)
```
