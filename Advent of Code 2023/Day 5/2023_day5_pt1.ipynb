```python
with open('2023_day5.csv', 'r') as f:
    s = f.readlines()
s = [x for x in s if x != '\n']
print(s)
```

```python
seeds = s[0].replace('\n','').split(':')[1].strip().split(' ')
seeds = [int(x) for x in seeds]
print(seeds)
```

```python
seed_soil_idx = s.index('seed-to-soil map:\n')
soil_fert_idx = s.index('soil-to-fertilizer map:\n')
fert_wate_idx = s.index('fertilizer-to-water map:\n')
wate_ligt_idx = s.index('water-to-light map:\n')
ligt_temp_idx = s.index('light-to-temperature map:\n')
temp_humi_idx = s.index('temperature-to-humidity map:\n')
humi_loca_idx = s.index('humidity-to-location map:\n')
```

```python
def build_map(start, end):
    map = []
    for i in range(start+1, end):
        map.append([int(x) for x in s[i].replace('\n','').split(' ')])
    return map
```

```python
seed_soil_maps = build_map(seed_soil_idx, soil_fert_idx)
soil_fert_maps = build_map(soil_fert_idx, fert_wate_idx)
fert_wate_maps = build_map(fert_wate_idx, wate_ligt_idx)
wate_ligt_maps = build_map(wate_ligt_idx, ligt_temp_idx)
ligt_temp_maps = build_map(ligt_temp_idx, temp_humi_idx)
temp_humi_maps = build_map(temp_humi_idx, humi_loca_idx)
humi_loca_maps = build_map(humi_loca_idx, len(s))
```

```python
def lkup(v, map):
    for r in map:
        if v >= r[1] and v <= r[1] + r[2]:
            return r[0] + (v - r[1])
    return v
```

```python
l = []
for seed in seeds:
    soil = lkup(seed, seed_soil_maps)
    fert = lkup(soil, soil_fert_maps)
    wate = lkup(fert, fert_wate_maps)
    ligt = lkup(wate, wate_ligt_maps)
    temp = lkup(ligt, ligt_temp_maps)
    humi = lkup(temp, temp_humi_maps)
    loca = lkup(humi, humi_loca_maps)
    l.append(loca)
print(min(l))
```
