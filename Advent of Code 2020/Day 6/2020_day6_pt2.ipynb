```python
with open('2020_day6.csv', 'r') as f:
    txt = f.read()
```

```python
groups = txt.split('\n\n')
print(groups)
```

```python
groups = [x.replace('\n', ',') for x in groups]
print(groups)
```

```python
counts = []
for g in groups:
    people = g.split(',')
    tests = people[0]
    for p in people[1:]:
        tests = set(tests).intersection(p)
    counts.append(len(tests))
print(counts)
```

```python
print(sum(counts))
```
