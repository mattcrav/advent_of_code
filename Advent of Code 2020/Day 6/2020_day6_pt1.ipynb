```python
from collections import Counter
```

```python
with open('2020_day6.csv', 'r') as f:
    txt = f.read()
```

```python
groups = txt.split('\n\n')
print(groups)
```

```python
groups = [x.replace('\n', '') for x in groups]
print(groups)
```

```python
counts = [len(Counter(x)) for x in groups]
print(counts)
```

```python
print(sum(counts))
```
