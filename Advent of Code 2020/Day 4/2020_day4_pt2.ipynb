```python
import pandas as pd
import numpy as np
import re
```

```python
with open('2020_day4.csv', 'r') as f:
    txt = f.read()
```

```python
passports = txt.split('\n\n')
print(passports)
```

```python
passports = [x.replace('\n', ' ') for x in passports]
passports = [x.split(' ') for x in passports]
print(passports)
```

```python
pass_dict = []
for p in passports:
    pass_dict.append({y.split(':')[0]:y.split(':')[1] for y in p})
print(pass_dict)
```

```python
df = pd.DataFrame(pass_dict)
df.head(50)
```

```python
def ecl_valid(ecl):
    if ecl is None:
        return 0
    if ecl in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']:
        return 1
    else:
        return 0
```

```python
def pid_valid(pid):
    if pid is None:
        return 0
    try:
        if len(pid) == 9 and int(pid) > -1:
            return 1
        else:
            return 0
    except:
        return 0
```

```python
def eyr_valid(eyr):
    if eyr is None:
        return 0
    if len(str(eyr)) ==  4 and int(eyr) >= 2020 and int(eyr) <= 2030:
        return 1
    return 0
```

```python
def hcl_valid(hcl):
    if hcl is None:
        return 0
    if str(hcl)[0] == '#' and re.match('[a-z0-9]', hcl[1:]):
        return 1
    return 0
    
```

```python
def byr_valid(byr):
    if byr is None:
        return 0
    if len(str(byr)) ==  4 and int(byr) >= 1920 and int(byr) <= 2002:
        return 1
    return 0
```

```python
def iyr_valid(iyr):
    if iyr is None:
        return 0
    if len(str(iyr)) ==  4 and int(iyr) >= 2010 and int(iyr) <= 2020:
        return 1
    return 0
```

```python
def cid_valid(cid):
    return 1
```

```python
def hgt_valid(hgt):
    if hgt is None:
        return 0
    if 'in' in str(hgt):
         v = int(hgt.split('in')[0]) 
         if v >= 59 and v <= 76:
             return 1
    if 'cm' in str(hgt):
        v = int(hgt.split('cm')[0]) 
        if v >= 150 and v <= 193:
             return 1
    return 0
```

```python
df['ecl_v'] = df['ecl'].apply(ecl_valid)
df['pid_v'] = df['pid'].apply(pid_valid)
df['eyr_v'] = df['eyr'].apply(eyr_valid)
df['hcl_v'] = df['hcl'].apply(hcl_valid)
df['byr_v'] = df['byr'].apply(byr_valid)
df['iyr_v'] = df['iyr'].apply(iyr_valid)
df['cid_v'] = df['cid'].apply(cid_valid)
df['hgt_v'] = df['hgt'].apply(hgt_valid)
df.head()
```

```python
df['valid'] = df[['ecl_v', 'pid_v', 'eyr_v', 'hcl_v', 'byr_v', 'iyr_v', 'cid_v', 'hgt_v']].sum(axis=1)
df.head()
```
