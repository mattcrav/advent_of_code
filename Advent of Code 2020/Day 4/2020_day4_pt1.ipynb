```python
with open('2020_day4.csv', 'r') as f:
    txt = f.read()
```

```python
passports = txt.split('\n\n')
print(passports)
```

```python
fields = [
    'byr:',
    'iyr:',
    'eyr:',
    'hgt:',
    'hcl:',
    'ecl:',
    'pid:',
    'cid:'
]
```

```python
valid = 0
for p in passports:
    matches = 0
    for f in fields:
        if f in p or f == 'cid:':
            matches += 1
    if matches == 8:
        valid += 1
```

```python
print(valid)
```
