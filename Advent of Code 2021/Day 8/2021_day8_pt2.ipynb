```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2021_day8.csv', header=None, names=['raw'])
df[['input', 'output']] = df['raw'].str.split('|', None, expand=True)
df['input'] = df['input'].str.strip().str.split(' ', None)
df['input'] = df['input'].apply(lambda x: [''.join(sorted(y)) for y in x])
df['output'] = df['output'].str.strip().str.split(' ', None)
df['output'] = df['output'].apply(lambda x: [''.join(sorted(y)) for y in x])
# df[[f'input{x}' for x in range(10)]] = df['input'].str.strip().str.split(' ', None, expand=True)
# df[[f'output{x}' for x in range(4)]] = df['output'].str.strip().str.split(' ', None, expand=True)
# for c in [f'output{x}' for x in range(4)]:
#     df[f'{c}'] = df[c].apply(lambda x: ''.join(sorted(x)))
# for c in [f'input{x}' for x in range(10)]:
#     df[f'{c}'] = df[c].apply(lambda x: ''.join(sorted(x)))
# for c in [f'output{x}' for x in range(4)]:
#     df[f'{c}len'] = df[c].apply(len)
df.head()
```

```python
df[1] = df['input'].apply(lambda x: next(y for y in x if len(y) == 2))
df[4] = df['input'].apply(lambda x: next(y for y in x if len(y) == 4))
df[7] = df['input'].apply(lambda x: next(y for y in x if len(y) == 3))
df[8] = df['input'].apply(lambda x: next(y for y in x if len(y) == 7))
df[9] = df.apply(lambda x: next(y for y in x['input'] if len(list(set(y) & set(x[4]))) == 4 and y != x[8] and y != x[4]), axis=1)

df['a'] = df.apply(lambda x: next(y for y in x[7] if y not in x[1]), axis=1)
df['g'] = df.apply(lambda x: next(y for y in x[9] if y not in x[4] + x['a']), axis=1)

df[3] = df.apply(lambda x: next(y for y in x['input'] if len(list(set(y) & set(x[1] + x['a'] + x['g']))) == 4 and y != x[8] and y != x[9] and len(y) == 5), axis=1)
df[0] = df.apply(lambda x: next(y for y in x['input'] if len(list(set(y) & set(x[1] + x['a'] + x['g']))) == 4 and y != x[8] and y != x[9] and len(y) == 6), axis=1)

df['d'] = df.apply(lambda x: next(y for y in x[8] if y not in x[0]), axis=1)
df['e'] = df.apply(lambda x: next(y for y in x[8] if y not in x[9]), axis=1)
df['b'] = df.apply(lambda x: next(y for y in x[8] if y not in x[1] + x['a'] + x['g'] + x['d'] + x['e']), axis=1)

df[6] = df.apply(lambda x: next(y for y in x['input'] if len(list(set(y) & set(x['a'] + x['b'] + x['g'] + x['d'] + x['e']))) == 5 and y != x[8]), axis=1)

df['c'] = df.apply(lambda x: next(y for y in x[8] if y not in x[6]), axis=1)
df['f'] = df.apply(lambda x: next(y for y in x[8] if y not in x['a'] + x['b'] + x['g'] + x['d'] + x['e'] + x['c']), axis=1)

df[2] = df.apply(lambda x: next(y for y in x['input'] if len(list(set(y) & set(x['a'] + x['c'] + x['g'] + x['d'] + x['e']))) == 5 and y != x[8]), axis=1)
df[5] = df.apply(lambda x: next(y for y in x['input'] if len(list(set(y) & set(x['a'] + x['b'] + x['g'] + x['d'] + x['f']))) == 5 and y != x[8] and y != x[6] and y != x[9]), axis=1)
df.head()
# df['a'] = np.where()
```

```python
df['key'] = df.apply(lambda x: {x[y]:str(y) for y in range(10)}, axis=1)
df.head()
```

```python
df['val'] = df.apply(lambda x: [x['key'][y] for y in x['output']],axis=1)
df['val'] = df.apply(lambda x: int(''.join(x['val'])),axis=1)
df.head()
```
