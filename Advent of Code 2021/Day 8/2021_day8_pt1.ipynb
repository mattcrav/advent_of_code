```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2021_day8.csv', header=None, names=['raw'])
df[['input', 'output']] = df['raw'].str.split('|', None, expand=True)
df[[f'input{x}' for x in range(10)]] = df['input'].str.strip().str.split(' ', None, expand=True)
df[[f'output{x}' for x in range(4)]] = df['output'].str.strip().str.split(' ', None, expand=True)
for c in [f'output{x}' for x in range(4)]:
    df[f'{c}len'] = df[c].apply(len)
df.head()
```

```python
# segs = {
#     0:6,
#     1:2,
#     2:5,
#     3:5,
#     4:4,
#     5:5,
#     6:6,
#     7:3,
#     8:7,
#     9:6
# }
segs = {
    2:1,
    4:4,
    3:7,
    7:8,
}
for c in [f'output{x}len' for x in range(4)]:
    df[f'{c}dig'] = df[c].map(segs)
df[[f'output{x}lendig' for x in range(4)]].head()
```
