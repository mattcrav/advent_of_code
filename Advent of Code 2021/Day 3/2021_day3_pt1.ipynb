```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2021_day3.csv', header=None, names=['val'], dtype='object')
df = df['val'].apply(lambda x: pd.Series(list(x)))
df.head(12)
```

```python
gamma = ''
epsilon = ''
for i in range(len(df.columns)):
    c = np.bincount(df[i])
    gamma += np.argmax(c).astype(str)
    epsilon += np.argmin(c).astype(str)
print('gamma: ', gamma, int(gamma, 2))
print('epsilon: ', epsilon, int(epsilon, 2))
print(int(gamma, 2) * int(epsilon, 2))
```
