```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2021_day3.csv', header=None, names=['val'], dtype='object')
df = df['val'].apply(lambda x: pd.Series(list(x)))
df.head(12)
```

```python
d = df.copy()
for i in range(len(df.columns)):
    c = np.bincount(d[i])
    # print(c)
    if c[0] == [c[1]]:
        m = 1
    else:
        m = np.argmax(c)
    d = d.loc[d[i].astype(int) == m]
    # print(d.head(12))
oxygen = d.apply(lambda x: ''.join(x.dropna().values.tolist()), axis=1).iloc[0]
print('oxygen: ', oxygen, int(oxygen, 2))
```
