```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2021_day9.csv', header=None, names=['raw'])
df = df['raw'].astype(str).str.split('', expand=True)
df[0] = 10
df[len(df.columns)-1] = 10
df = df.astype(int)
wall = pd.DataFrame([[10 for x in range(len(df.columns))]])
df = pd.concat([wall, df, wall])
df.head()
```

```python
n = df.to_numpy()
```

```python
low_points = []
for i in range(2,len(n)):
    for j in range(2,len(n[0])):
        v = n[i-1:i,j-1:j].sum()
        u = n[i-2:i-1,j-1:j].sum()
        d = n[i:i+1,j-1:j].sum()
        l = n[i-1:i,j-2:j-1].sum()
        r = n[i-1:i,j:j+1].sum()
        if v < u and v < d and v < l and v < r:
            low_points.append([i-1,j-1])
print(low_points)
```

```python
b = np.where((n == 9) | (n == 10), 0, 1)
print(b)
sizes = []
for l in low_points:
    to_explore = [l]
    size = 0
    while len(to_explore) > 0:
        i = to_explore.pop()
        if b[i[0], i[1]] == 1:
            size += 1
        u = b[i[0]-1, i[1]].sum()
        d = b[i[0]+1, i[1]].sum()
        l = b[i[0], i[1]-1].sum()
        r = b[i[0], i[1]+1].sum()
        b[i[0], i[1]] = 0
        if u == 1:
            to_explore.append([i[0]-1, i[1]])
        if d == 1:
            to_explore.append([i[0]+1, i[1]])
        if l == 1:
            to_explore.append([i[0], i[1]-1])
        if r == 1:
            to_explore.append([i[0], i[1]+1])
        # size += u + d + l + r
    sizes.append(size)
sizes.sort(reverse=True)
x = np.array(sizes[:3])
print(x)
print(x.prod())
```
