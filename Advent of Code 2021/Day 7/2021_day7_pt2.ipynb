```python
import pandas as pd
import numpy as np
```

```python
with open('2021_day7.csv', 'r') as f:
    l = f.readlines()[0].split(',')
    l = np.array([int(x) for x in l])
print(l)
```

```python
r = []
for i in range(max(l)+1):
    f = sum([np.arange(x+1).sum() for x in np.abs(l - i)])
    # print(f)
    r.append(f)
print(min(r))
```
